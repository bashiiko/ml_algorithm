# titanicデータセットの前処理

```python
# 欠損値の確認
def check_null(df): 
        null_val = df.isnull().sum()
        percent = 100 * df.isnull().sum()/len(df)
        null_table = pd.concat([null_val, percent], axis=1)
        null_table = null_table.rename(
        columns = {0 : '欠損数', 1 : '%'})
        return null_table
 
```

```python
check_null(data)
```

```python
# 分析に使用するデータの欠損地を埋める
data["Age"] = data["Age"].fillna(data["Age"].median())
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])

# "Fare"が欠損している乗客のクラス
pclass = data[data["Fare"].isnull()]["Pclass"]
data["Fare"] = data["Fare"].fillna(data[data["Pclass"]==int(pclass)]["Fare"].median())

check_null(data)
```

```python
# データの前処理

data["Sex"][data["Sex"] == "male"] = 0
data["Sex"][data["Sex"] == "female"] = 1

data["Embarked"][data["Embarked"] == "S" ] = 0
data["Embarked"][data["Embarked"] == "C" ] = 1
data["Embarked"][data["Embarked"] == "Q"] = 2

data = data.drop("Name", axis=1)
data = data.drop("Ticket", axis=1)
data = data.drop("Cabin", axis=1)
data = data.drop("PassengerId", axis=1)

data.to_csv("../dataset/titanic/data_preprocess.csv")
```

# 自作決定木
【参考】http://darden.hatenablog.com/?page=1485345164
【参考】http://hktech.hatenablog.com/entry/2018/10/05/004235

```python
import numpy as np

class Node(object):
    def __init__(self, criterion="gini", max_depth = 3):
        self.criterion = criterion
        self.left = None
        self.right = None
        self.max_depth = max_depth
        self.depth = None
        self.nSample = None
        self.label = None
        self.impurity = None # 不純度
        self.f_index = None
        self.thrs = None
    
    def make_child_node(self, depth, features, target, nClass):
        # Input:
        #  depth[1×1]  current depth
        #  features[nSample×nDim]  explanatory variable
        #  target[nSample×1]  objective variable
        #  nClass[1×1]  number of classes(objective variable)
        
        self.depth = depth
        self.nSample = features.shape[0]
        nDim = features.shape[1]
        
        # ノード内のクラスが単一の場合終了
        if len(np.unique(target)) == 1:
            self.impurity = 0
            return

        self.impurity = self.calc_criterion(target)
        self.label = np.argmax(np.bincount(target))
        
        max_gain, self.thrs, self.f_index = self.search_split_point(features, target)

        if self.depth == self.max_depth:
            return
        
        left_idx = features[:, self.f_index] <= self.thrs
        right_idx = features[:, self.f_index] > self.thrs
    
        self.left = Node(self.criterion, self.max_depth)
        self.right = Node(self.criterion, self.max_depth)
        
        self.left.make_child_node(depth+1, features[left_idx], target[left_idx], nClass)
        self.right.make_child_node(depth+1, features[right_idx], target[right_idx], nClass)
        
    def search_split_point(self, features, target):
        best_thrs = None # 分割に最適なしきい値
        best_f = None    # 分割に最適な説明変数
        max_gain = 0     # gainの最大値
        
        # 特徴量の順番をランダムにする
        f_loop_order = np.random.permutation(features.shape[1]).tolist()
        for idx in f_loop_order:
            f_selected = features[:,idx]
            for thrs in f_selected:
                left_idx = features[:, idx] <= thrs
                right_idx = features[:, idx] > thrs
                gain = self.calc_gain(target, target[left_idx], target[right_idx])
                if max_gain < gain:
                    max_gain = gain
                    best_thrs = thrs
                    best_f = idx
        
        return max_gain, best_thrs, best_f
                
        
    def calc_criterion(self, target):
        classes = np.unique(target) # 現在のノードに属するデータのクラス一覧
        N = len(target)
        
        if self.criterion == "gini":
            I = 1
            for c in classes:
                p = float(len(target[target == c])) / N
                I -= p ** 2 
        
        elif self.criterion == "entropy":
            I = 0
            for c in classes:
                p = float(len(target[target == c])) / N
                I -= p * np.log2(p)
        
        return I
                
    def calc_gain(self, target_p, target_cl, target_cr):

        cri_p = self.calc_criterion(target_p)
        cri_cl = self.calc_criterion(target_cl)
        cri_cr = self.calc_criterion(target_cr)

        gain = cri_p - (float(len(target_cl))/len(target_p)) * cri_cl \
                - (float(len(target_cr))/len(target_p)) * cri_cr
        
        return gain
    
    def predict(self, sample):
        if self.f_index == None or self.depth == self.max_depth:
            return self.label
        else:
            if sample[self.f_index] <= self.thrs:
                return self.left.predict(sample)
            elif sample[self.f_index] > self.thrs:
                return self.right.predict(sample)
        
        
```

```python
class DecisionTree(object):
    def __init__(self, criterion="gini", max_depth=3):
        self.tree = None
        self.criterion = criterion
        self.max_depth = max_depth
    
    def fit(self, features, target):
        self.tree = Node(self.criterion, self.max_depth)
        self.tree.make_child_node(depth=0, features=features, target=target, nClass=len(np.unique(target)))
    
    def predict(self, samples):
        pred = []
        for s in samples:
            pred.append(self.tree.predict(s))
            
        return np.array(pred)
            
    def score(self, sample, target):
        return sum(self.predict(sample) == target) / float(len(target))
        
```

# scikit-learnのライブラリを使用した場合と結果を比較

```python
from sklearn import tree
import pandas as pd

titanic= pd.read_csv("../dataset/titanic/data_preprocess.csv")

feature_names = ['Pclass',	'Sex', 'Age', 'SibSp', 'Parch', 'Fare',	'Embarked']
class_names = ['Dead', 'Survived']
data = titanic[feature_names]
target = titanic['Survived']

## original model
clf_m = DecisionTree(criterion="gini", max_depth=3)
clf_m.fit(data.values,target.values)

## scikit-learn model
clf_s = tree.DecisionTreeClassifier( criterion = 'gini', max_depth=3 )
clf_s = clf_s.fit(data, target)

## predict
pred_score = clf_m.score(data.values, target.values)
predicted = clf_s.predict(data)

## output
#print('=============== predicted ===============')
#print(predicted)
#print('============== correct_ans ==============')
#print(target)
#print('=============== id_rate =================')
print("original model's score")
print(pred_score)

print("scikit-learn model's score")
print(sum(predicted == target) / len(target))
```

```python
# 交差検証
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=None)

score_m = []
score_s = []
for train_index, test_index in kfold.split(data.values, target.values):
    clf_m.fit(data.values[train_index],target.values[train_index])
    clf_s.fit(data.values[train_index],target.values[train_index])
    
    
    predicted = clf_s.predict(data.values[test_index])
    
    score_m.append(clf_m.score(data.values[test_index], target.values[test_index]))
    score_s.append(sum(predicted == target[test_index]) / len(target[test_index]))
       
print("original model's score")
#score_m
print(np.mean(score_m))
print("scikit-learn model's score")
print(np.mean(score_s))
```

```python
import pydotplus
from sklearn.externals.six import StringIO
from IPython.display import Image

## visuzlization
dot_data = StringIO()
dot_data = tree.export_graphviz(clf_s, out_file = None, feature_names = feature_names,
                           class_names = class_names, filled = True, rounded = True)


#tree.export_graphviz(clf, out_file=dot_data)
graph = pydotplus.graph_from_dot_data(dot_data)

graph.write_png("./result/graph.png")

Image(graph.create_png())
```
